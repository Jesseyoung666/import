```python
import pandas as pd

# Load the dataset
file_path = 'data/import export of all countries.csv'
data = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Information:")
data.info(verbose=True, memory_usage="deep")

print("\nFirst Few Rows of the Dataset:")
print(data.head())
```

```python
# 数据质量检查
```

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)

# Basic statistics for numerical columns
print("\nBasic Statistical Overview:")
print(data.describe())

# Filtering out non-numeric columns for outlier detection
numeric_data = data.select_dtypes(include=['number'])

# Checking for potential outliers in the numeric columns
print("\nPotential Outliers Check in Numeric Columns:")
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))).sum()
print(outliers)
```

```python
#主要贸易国家：确定进出口量最大的国家。
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 选择一个特定年份进行分析，例如2022年
year = '2022 [YR2022]'

# 过滤出相关的进出口数据
exports_data = data[data['Series Name'].str.contains('exports', case=False, na=False)]
imports_data = data[data['Series Name'].str.contains('imports', case=False, na=False)]

# 将相关年份的数据转换为数值类型
exports_data[year] = pd.to_numeric(exports_data[year], errors='coerce')
imports_data[year] = pd.to_numeric(imports_data[year], errors='coerce')

# 计算每个国家在所选年份的总出口和进口
exports_sum = exports_data.groupby('Country Name')[year].sum()
imports_sum = imports_data.groupby('Country Name')[year].sum()

# 找到出口和进口最大的国家
max_export_country = exports_sum.idxmax()
max_import_country = imports_sum.idxmax()

# 绘制条形图展示
plt.figure(figsize=(12, 6))
plt.bar(['Max Export Country', 'Max Import Country'], [exports_sum[max_export_country], imports_sum[max_import_country]])
plt.title(f'Major Trade Countries in {year}')
plt.ylabel('Trade Value (Percentage)')
plt.show()
```

```python
#贸易额分布：展示不同国家的贸易额分布情况。
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 选择一个特定年份进行分析，例如最近的年份，这里我们假设为2021年
year = '2021 [YR2021]'

# 过滤出相关的进出口数据
exports_data = data[data['Series Name'].str.contains('exports', case=False, na=False)]
imports_data = data[data['Series Name'].str.contains('imports', case=False, na=False)]

# 将相关年份的数据转换为数值类型
exports_data[year] = pd.to_numeric(exports_data[year], errors='coerce')
imports_data[year] = pd.to_numeric(imports_data[year], errors='coerce')

# 计算每个国家在所选年份的总出口和进口
exports_sum = exports_data.groupby('Country Name')[year].sum()
imports_sum = imports_data.groupby('Country Name')[year].sum()

# 计算总贸易额（出口加进口）
total_trade = exports_sum + imports_sum

# 取出总贸易额最高的前10个国家
top_countries = total_trade.nlargest(10)

# 绘制条形图展示
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.title(f'Trade Value Distribution for Top 10 Countries in {year}')
plt.ylabel('Total Trade Value (Percentage)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

```python
#贸易平衡分析：计算每个国家的贸易平衡（出口减去进口），以识别贸易顺差或逆差国家。
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 选择一个特定年份进行分析，例如最近的年份，这里我们假设为2021年
year = '2021 [YR2021]'

# 过滤出相关的进出口数据
exports_data = data[data['Series Name'].str.contains('exports', case=False, na=False)]
imports_data = data[data['Series Name'].str.contains('imports', case=False, na=False)]

# 将相关年份的数据转换为数值类型
exports_data[year] = pd.to_numeric(exports_data[year], errors='coerce')
imports_data[year] = pd.to_numeric(imports_data[year], errors='coerce')

# 计算每个国家在所选年份的总出口和进口
exports_sum = exports_data.groupby('Country Name')[year].sum()
imports_sum = imports_data.groupby('Country Name')[year].sum()

# 计算贸易平衡（出口减去进口）
trade_balance = exports_sum - imports_sum

# 取出贸易平衡的前10个国家（正值表示贸易顺差，负值表示贸易逆差）
top_trade_balances = trade_balance.nlargest(10)
bottom_trade_balances = trade_balance.nsmallest(10)

# 绘制条形图展示
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
top_trade_balances.plot(kind='bar', color='green')
plt.title(f'Top 10 Countries with Trade Surplus in {year}')
plt.ylabel('Trade Balance (Percentage)')
plt.xlabel('Country')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
bottom_trade_balances.plot(kind='bar', color='red')
plt.title(f'Top 10 Countries with Trade Deficit in {year}')
plt.ylabel('Trade Balance (Percentage)')
plt.xlabel('Country')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 选择要分析的国家，例如中国
country_name = 'China'

# 过滤出相关的进出口数据
exports_data = data[(data['Country Name'] == country_name) & 
                    data['Series Name'].str.contains('exports', case=False, na=False)]
imports_data = data[(data['Country Name'] == country_name) & 
                    data['Series Name'].str.contains('imports', case=False, na=False)]

# 提取年份列
year_columns = exports_data.columns[4:]

# 将相关年份的数据转换为数值类型
exports_values = pd.to_numeric(exports_data.iloc[0, 4:], errors='coerce')
imports_values = pd.to_numeric(imports_data.iloc[0, 4:], errors='coerce')

# 计算总贸易额（出口加进口）
total_trade_values = exports_values + imports_values

# 绘制时间序列图
plt.figure(figsize=(12, 6))
plt.plot(year_columns, total_trade_values, marker='o')
plt.title(f'Trade Value Time Series Analysis for {country_name}')
plt.xlabel('Year')
plt.ylabel('Total Trade Value (Percentage)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 进出口比较：对比不同国家的进口和出口情况。
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 选择要分析的国家和年份
countries = ['China', 'United States', 'Germany', 'Japan', 'India']  # 示例国家列表
year = '2021 [YR2021]'  # 示例年份

# 过滤出相关的进出口数据
exports_data = data[data['Series Name'].str.contains('exports', case=False, na=False)]
imports_data = data[data['Series Name'].str.contains('imports', case=False, na=False)]

# 将相关年份的数据转换为数值类型
exports_data[year] = pd.to_numeric(exports_data[year], errors='coerce')
imports_data[year] = pd.to_numeric(imports_data[year], errors='coerce')

# 获取所选国家的进出口数据
exports = exports_data[exports_data['Country Name'].isin(countries)].groupby('Country Name')[year].sum()
imports = imports_data[imports_data['Country Name'].isin(countries)].groupby('Country Name')[year].sum()

# 绘制堆叠条形图进行比较
fig, ax = plt.subplots(figsize=(12, 6))
exports.plot(kind='bar', ax=ax, color='skyblue', position=1, width=0.4, label='Exports')
imports.plot(kind='bar', ax=ax, color='lightcoral', position=0, width=0.4, label='Imports')

plt.title(f'Import and Export Comparison for Selected Countries in {year}')
plt.ylabel('Trade Value (Percentage)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# 地理分布图：使用地图展示不同国家和地区的贸易额。
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# 加载世界地图
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# 选择要分析的年份
year = '2021 [YR2021]'

# 过滤出相关的贸易数据
# 这里我们简化，只考虑出口数据作为贸易额的代表
exports_data = data[data['Series Name'].str.contains('exports', case=False, na=False)]
exports_data[year] = pd.to_numeric(exports_data[year], errors='coerce')
exports_sum = exports_data.groupby('Country Name')[year].sum()

# 将贸易数据合并到世界地图数据中
world = world.merge(exports_sum, left_on='name', right_on='Country Name', how='left')

# 绘制地图
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax)
world.plot(column=year, ax=ax, legend=True,
           legend_kwds={'label': "Export Value by Country",
                        'orientation': "horizontal"})

plt.title(f'World Trade Export Value Distribution in {year}')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 选择要分析的几个国家和年份范围
countries = ['Afghanistan', 'Albania', 'Brazil']  # 示例国家列表
years = ['2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]']  # 示例年份范围

# 创建一个空的DataFrame来存储数据
comparison_data = pd.DataFrame(index=years)

# 提取并存储每个国家的出口数据
for country in countries:
    country_data = data[(data['Country Name'] == country) &
                        (data['Series Name'].str.contains('exports', case=False, na=False))]
    comparison_data[country] = country_data.iloc[0][years].replace('..', pd.NA).astype(float)

# 绘制折线图
comparison_data.plot(kind='line', figsize=(10, 6))
plt.title('Export Data Comparison Among Countries')
plt.xlabel('Year')
plt.ylabel('Exports (% of commercial service exports)')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 假设 data 是您的 DataFrame

# 提取所有年份的列
year_columns = data.columns[4:]

# 初始化一个字典来存储每个年份的平均贸易额
average_trade_by_year = {}

# 计算每个年份的平均贸易额
for year in year_columns:
    # 将非数值替换为NaN，然后计算平均值
    average_trade = pd.to_numeric(data[year].replace('..', pd.NA), errors='coerce').mean()
    average_trade_by_year[year] = average_trade

# 转换为DataFrame
average_trade_df = pd.DataFrame(list(average_trade_by_year.items()), columns=['Year', 'Average Trade Value'])

# 绘制折线图
plt.figure(figsize=(12, 6))
plt.plot(average_trade_df['Year'], average_trade_df['Average Trade Value'], marker='o')
plt.title('Average Trade Value Over Years for All Countries')
plt.xlabel('Year')
plt.ylabel('Average Trade Value (Percentage)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 假设 data 是您的 DataFrame

# 选择一些年份作为特征
selected_years = ['2000 [YR2000]', '2010 [YR2010]', '2020 [YR2020]']

# 准备聚类数据
cluster_data = data[selected_years].replace('..', pd.NA)
cluster_data_dropna = cluster_data.dropna()  # 移除任何含NaN的行
cluster_data_dropna = cluster_data_dropna.astype(float)  # 确保所有数据为数值类型

# 应用K-means聚类
kmeans = KMeans(n_clusters=3)
kmeans.fit(cluster_data_dropna)

# 在原始数据集中创建一个新列用于存储聚类结果
data['Cluster'] = np.nan

# 为参与聚类的行更新聚类结果
data.loc[cluster_data_dropna.index, 'Cluster'] = kmeans.labels_

# 绘制聚类结果的可视化
# 仅绘制参与聚类的数据点
plt.figure(figsize=(8, 6))
plt.scatter(data.loc[cluster_data_dropna.index, '2000 [YR2000]'], 
            data.loc[cluster_data_dropna.index, '2020 [YR2020]'], 
            c=data.loc[cluster_data_dropna.index, 'Cluster'], 
            cmap='viridis')
plt.title('Cluster of Countries based on Trade Data')
plt.xlabel('Trade Value in 2000')
plt.ylabel('Trade Value in 2020')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 假设 data 是您的 DataFrame

# 选择所有年份的列
year_columns = data.columns[4:]

# 准备相关性分析的数据
# 将非数值 ('..') 替换为 NaN
correlation_data = data[year_columns].replace('..', np.nan)

# 将数据转换为数值类型
correlation_data = correlation_data.astype(float)

# 计算相关系数矩阵
correlation_matrix = correlation_data.corr()

# 使用Seaborn绘制热力图
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Import and Export Data')
plt.show()
```
